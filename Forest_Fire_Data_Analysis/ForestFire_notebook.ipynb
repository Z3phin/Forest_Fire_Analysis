## Imported Libraries

```python
# Data Manipulation
import numpy as np
import pandas as pd

# Data Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Model Evaluation
from sklearn.decomposition import PCA
```

# Data Description

The dataset includes 243 instances that cover forest fires across two regions of Algeria.

The dataset includes 11 attribues and 1 output attribue ("Target")
The 243 instances have been classified into "fire" and "not fire" classes.

Attribute Information (Name : Description):
1. Day : day of the month 
2. Month : month of the year ('june' to 'september')
3. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
4. RH : Relative Humidity in %: 21 to 90
5. Ws : Wind speed in km/h: 6 to 29
6. Rain: total day in mm: 0 to 16.8
7. Fine Fuel Moisture Code (FFMC) : index from the FWI system: 28.6 to 92.5
8. Duff Moisture Code (DMC) :  index from the FWI system: 1.1 to 65.9
9. Drought Code (DC) :  index from the FWI system: 7 to 220.4
10. Initial Spread Index (ISI) : index from the FWI system: 0 to 18.5
11. Buildup Index (BUI) : index from the FWI system: 1.1 to 68
12. Fire Weather Index (FWI) : Index: 0 to 31.1
13. Target : two classes, namely "Fire" and "not Fire"

## Loading Data

```python
# Read data into data frame
fire_frame = pd.read_csv("ForestFire.csv")

# Add column titles 
fire_frame.columns = ["Day", "Month", "Temp", "RH", "Ws", "Rain", "FFMC", "DMC", "DC", "ISI", "BUI", "FWI", "Target"]

# Splitting the data into feature frame and label frame
feature_frame = fire_frame.drop(['Target'], axis=1)
label_frame = fire_frame['Target']
```

# Exploratory Analysis

## Pie Chart

```python
plt.pie(label_frame.value_counts(),
        labels = ["Fire", "Not Fire"],
        autopct="%.2f%%",
        shadow=True)
plt.title("Proportion of Fire and Not Fire")
plt.show()
```

## Principal Component Analysis

```python
# Standardisation of data

standardised_feature_frame = (feature_frame-feature_frame.mean())/feature_frame.std()

# PCA Model

pca_model = PCA()
pca_transformed_frame = pca_model.fit_transform(standardised_feature_frame)

# PCA visualisation

colours = label_frame.copy()
colours.replace("notfire", "b", inplace=True)
colours.replace("fire", "r", inplace=True)

plt.figure()
plt.figure(figsize = (5,5))
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Principal Component 1', fontsize = 14)
plt.ylabel('Principal Component 2', fontsize = 14)
plt.title("Principal Component Analysis of Forest Fire Dataset", fontsize = 16)
plt.scatter(pca_transformed_frame[:, 0], pca_transformed_frame[:, 1], c = colours)
plt.show()
```

There's a clear distinction between fire and not a fire. It may be suitable for predictions to be made from.
